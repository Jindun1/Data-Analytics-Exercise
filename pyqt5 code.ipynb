```python
# Create 3 dataframe
import pandas as pd
data = [['Tom', 10], ['Nick', 12], ['Jeff', 11]] 
  
# Create the pandas DataFrame 
df1 = pd.DataFrame(data, columns = ['Name', 'Age']) 
data = [['Linda', 10], ['Peter', 5], ['Lily', 16]] 
  
# Create the pandas DataFrame 
df2 = pd.DataFrame(data, columns = ['Name', 'Age'])
data = [['Sophie', 6], ['Chris', 15], ['Lucy', 7]] 
  
# Create the pandas DataFrame 
df3 = pd.DataFrame(data, columns = ['Name', 'Age'])
```

```python
import sys
import pandas as pd
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *


class Form(QDialog):
    
    value_selected=''
    value_selected2=''
    df=pd.DataFrame(columns=['Class'])
    def __init__(self, parent=None):
        super(Form, self).__init__(parent)

        button2 = QPushButton('Upload a file')
        button2.clicked.connect(self.openfile)
        button2.clicked.connect(self.creat_table_show)
        
        
        
        button1 = QPushButton('Go')
        button1.clicked.connect(lambda: self.on_button(df1,df2,df3))
        
        layout = QVBoxLayout()
        
        
        self.btn1=QRadioButton('Original data')
        # Default value
        self.btn1.setChecked(True)
        self.btn1.toggled.connect(lambda :self.btnstate(self.btn1))
        

        self.btn2 = QRadioButton('Undersampling data')
        self.btn2.toggled.connect(lambda: self.btnstate(self.btn2))
        
        
        self.btn3 = QRadioButton('Oversampling data')
        self.btn3.toggled.connect(lambda: self.btnstate(self.btn3))
        
        
        self.label=QLabel("Select Attribute to analyse:")
        self.cb=QComboBox()
        self.cb.addItems(Form.df.columns.tolist())
        self.cb.currentIndexChanged.connect(self.selectionchange)
          
        layout.addWidget(button2)
        layout.addWidget(self.btn1)
        layout.addWidget(self.btn2)
        layout.addWidget(self.btn3)
        
        layout.addWidget(self.label)
        layout.addWidget(self.cb)
        
        layout.addWidget(button1)
        
        self.resize(300,90)
        self.setLayout(layout)
        self.setWindowTitle('RC Analysis')

    def selectionchange(self,i):
        
        print('current index',i,'selection changed',self.cb.currentText())
        Form.value_selected2=self.cb.currentText()
   
    def btnstate(self,btn):
    
        if btn.text()=='Original data':
            if btn.isChecked()==True:
                print(btn.text()+"is selected")
                Form.value_selected=btn.text()
            

        if btn.text()=="Undersampling data":
            if btn.isChecked() == True:
                print(btn.text() + "is selected")
                Form.value_selected=btn.text()
                
            
        if btn.text()=="Oversampling data":
            if btn.isChecked() == True:
                print(btn.text() + "is selected")
                Form.value_selected=btn.text()
            
 
    def on_button(self, df1,df2,df3):
        writer = pd.ExcelWriter('ds result.xlsx', engine='xlsxwriter')

        # Write each dataframe to a different worksheet.
        df1.to_excel(writer, sheet_name='original')
        df2.to_excel(writer, sheet_name='undersampling')
        df3.to_excel(writer, sheet_name='oversampling')

        writer.save()
    
    def openfile(self):
        openfile_name = QFileDialog.getOpenFileName(self,'Select File','','Excel files(*.xlsx , *.xls)')
        global path_openfile_name
        path_openfile_name = openfile_name[0]


    def creat_table_show(self):
        if len(path_openfile_name) > 0:
            input_table = pd.read_excel(path_openfile_name)    
            Form.df=input_table
            self.cb.clear()       # delete all items from comboBox
            self.cb.addItems(input_table) 
            
if __name__ == '__main__':
    
    app = QApplication(sys.argv)
    form = Form()
    form.show()

    sys.exit(app.exec_())
```
