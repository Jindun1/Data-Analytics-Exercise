```python
# -*- coding: utf-8 -*-
"""
Created on Wed May 22 16:33:20 2019

@author: ooezuemerzifon
"""

#####Collect all the file names####
import pandas as pd
from os import listdir
from os.path import isfile, join
mypath='K:/Lyth/Oemer/24_vorhandene_Felder/Decompressed/' # Define the path of the folder which contains all the files
rawdata = [f for f in listdir(mypath) if isfile(join(mypath, f))]

# Save all the files' name in the dataframe df
df=pd.DataFrame(data=rawdata,columns=['Files'])

# Define a new column 'Tables' to save all the cleaned file names (without digits)
df['Tables'] = ''
for i in range(len(df)):
    #print(df.loc[i, 'Files'])
    A = ''
    B = ''
    A = df.at[i, 'Files'].split('_')
    for a in A:
        if (a[0].isdigit() == False):
            B = B + '_' + a
    df.at[i, 'Tables'] = B[1:]
    
#####Create a new txt file to show the column names of each file####
# the structure of the txt file would be:
#    TableName1: Column names Table1
#    TableName2: Column names Table2
#    ...

processed = list()
with open('Data.txt', 'w+') as writeFile:
    for i in range(len(df)):
        if df.at[i, 'Tables'] not in processed:
            with open('Decompressed/'+df.at[i, 'Files']) as readFile:
                processed.append(df.at[i, 'Tables'])
                for j in readFile.readline().split('#|#'):
                    writeFile.write(df.at[i, 'Tables'] + '_' + j.replace('ï»¿', '')+'\n')
                    
#####Create a new txt file to show the row names of each file####
# the structure of the txt file would be:
#    TableName1: Row names Table1
#    TableName2: Row names Table2
#    ...

processed = list()
with open('Data.txt', 'w+') as writeFile:
    for i in range(len(df)):
        if df.at[i, 'Tables'] not in processed:
            with open('Decompressed/'+df.at[i, 'Files']) as readFile:
                processed.append(df.at[i, 'Tables'])
                result=[]                                           
                for x in readFile.readlines():
                    result.append(x.split('#|#')[0])
                for j in result:
                    writeFile.write(df.at[i, 'Tables'] + '_' + j.replace('ï»¿', '')+'\n') 
                    
#####Create a new txt file to show the row number of each file####
# the structure of the txt file would be:
#    TableName1:  Number of Rows Table1
#    TableName2: Number of Rows Table2
#    ...

processed = list()
with open('Data.txt', 'w+') as writeFile:
    for i in range(len(df)):
        if df.at[i, 'Tables'] not in processed:
            with open('Decompressed/'+df.at[i, 'Files']) as readFile:
                processed.append(df.at[i, 'Tables'])                                          
                lines=readFile.readlines()
                number=len(lines)
                writeFile.write(df.at[i, 'Tables'] + ':' + number +'\n')                    
```
